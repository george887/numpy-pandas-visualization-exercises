```python
# Use the iris database to answer the following quesitons:
iris = sns.load_dataset('iris')
iris
```

```python
#1A. What does the distribution of petal lengths look like? 150 rows and 5 columns
iris.shape
```

```python
#1B. Is there a correlation between petal length and petal width? Yes as petal length increases so does width
sns.relplot(x="petal_length", y="petal_width", data = iris)
plt.title('Scatterplot of petal width and length')
```

```python
#1 C. Would it be reasonable to predict species based on sepal width and sepal length?
# Yes, we can see a clearly how setosa petals compare to versicolor and virginica. 
# For the most part, you would be able to tell the difference of versicolor and virginica but it could become
#difficult to predict the species as they merge together.
sns.relplot(x="petal_length", y="petal_width", hue='species', data = iris)
plt.title('Scatterplot of petal width and length')
```

```python
#1D. Which features would be best used to predict species? Petal width vs Petal length. Sepal length could be
# used to distinguish Setosa species but can become difficult with versicolor and virginica as they overlap
sns.relplot(x="sepal_length", y="sepal_width", hue='species', data = iris)
plt.title('Scatterplot of seapal width and length')
```

```python
#3. Load the swiss dataset and read it's documentation. Create visualizations to answer
# the following questions:
swiss = data('swiss')
swiss.head()
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is
# Catholic. (Choose a cutoff point for what constitutes catholic)
swiss['is_catholic'] = swiss.Catholic > 51.0
data('swiss', show_doc = True)
```

```python
#3B. Does whether or not a province is Catholic influence fertility?
# Yes, there is a correlation of being Catholic and being more fertile
sns.relplot(x='Catholic', y='Fertility', hue='is_catholic', data=swiss)
plt.title('Scatterplot of Catholic and Fertility')
```

```python
# Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most 
# popular items and the revenue produced by each.
from env import host, password, user
def get_db_url(db, user=user, host=host, password=password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
chipotle_sql_query = '''
                     SELECT *
                     FROM orders;
                     '''
orders = pd.read_sql(chipotle_sql_query, get_db_url('chipotle'))
orders
top_four = orders.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
top_four
```

```python
# This way I can also see the revenue produced by item as well as item.
orders['item_total'] = orders.quantity * orders.item_price
orders.groupby('item_name').item_total.sum().sort_values(ascending=False).head(4)
```
