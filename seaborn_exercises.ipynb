```python
# Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most 
# popular items and the revenue produced by each.
# creating a get_db_url function to take in db, user, host, password. Accessing the chipotle dataset select *
from env import host, password, user
def get_db_url(db, user=user, host=host, password=password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
chipotle_sql_query = '''
                     SELECT *
                     FROM orders;
                     '''
```

```python
# Creating orders variable. Using the pd.read_sql function to create the orders DF. in the chipotle_sql_query.
# Using the get_db_url variable to the chipotle dataset
orders = pd.read_sql(chipotle_sql_query, get_db_url('chipotle'))
orders.head()
```

```python
# creating a new colum item_price in the order df. Using the item price column as a string to replacing the $
# to make a int I can do mathmatical operations and setting as a float to account for the change. 
orders['item_price'] = orders.item_price.str.replace('$','').astype(float)
orders.head()
```

```python
#Top 4 popular items
#creating a top four variable in my orders DF. grouping by the item name. Get the quantity and summing. sort 
# values ascending to see the highest ones and head(4) for top 4
top_four = orders.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
top_four
```

```python
#Created a revenue variable in the orders df. grouping by item name, getting the item price and aggregating the
# sum
revenue = orders.groupby('item_name').item_price.agg('sum')
revenue.head()
```

```python
#Merging the top four items and revenue  on an inner join to create a new df
top_four_revenue = pd.merge(top_four, revenue, left_on ='item_name', right_on ='item_name', how ='inner')
top_four_revenue
```

```python
#plotting the top four popular items at Chipotle with revenue
#Adjusting the figure size
plt.figure(figsize=(10, 5))
#adding plot title
plt.title('Revenue for Top 4 Popular Items at Chipotle')
#created a sns barplot. using x as **top_four_revenue.index**
sns.barplot(x = top_four_revenue.index, y = 'item_price',  data = top_four_revenue)

plt.xlabel('Item')
plt.ylabel('Revenue')
```

```python
#Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the
#individual subject's reaction times and a more prominant line showing the average change in reaction time.
sleepstudy = data('sleepstudy')
data('sleepstudy', show_doc = True)
```

```python
sleepstudy.head()
```

```python
# created a df avg_reaction. in the sleepstudy dataset. grouping by days and taking the average
avg_reaction = sleepstudy.groupby('Days').mean()
avg_reaction.head()
```

```python
plt.figure(figsize=(12, 8))

plt.title('Reaction time with days of sleep deprivation')
# Accessing the avg_reaction df. .Reaction column. Added a wider linewidth to see better and line style
plt.plot(avg_reaction.Reaction, linewidth = 5, linestyle="--")
plt.xlabel('Days')
plt.ylabel('Reaction')
# Created the sns plot on top of the plt of avg reaction time. palette = 'Paired' gives each line a unique color
sns.lineplot('Days', 'Reaction', hue = 'Subject', data = sleepstudy, palette = 'Paired')
#Have to move legend as it is on the data
plt.legend(loc='best', bbox_to_anchor=(1.02,.9))
```
