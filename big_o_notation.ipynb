#5 Title your chart "Big O Notation"
#### Label your x axis "Elements"
#### Label your y axis "Operations"
#### Label your curves or make a legend for the curves
#### Use LaTex notation where possible
#### Curves to graph 
#### y = 0x+1 and label the curve "O(1)"
#### y = log(x) and label the curve "O(log n)"
#### y = x and label the curve "O(n)"
#### y = x * log(x) and label it "O(n log n)"
#### y = x^2 and label it "O(n^2)"
#### y = 2^x and label it "O(2^n)"
#### y = n! and label it "O(n!)"
#### y = n^n and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(1,5))

y1 = [((0*n) + 1) for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]


plt.plot(x,y1, label = "O(1)")
plt.plot(x,y2, label = "O(log n)")
plt.plot(x,y3, label = "O(n")
plt.plot(x,y4, label = "O(n log n)")
plt.legend()
plt.title('Big O Notations 1st four equations')
plt.xlabel('Elements')
plt.ylabel('Operations')
```

```python
x = list(range(0,3))
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n ** n for n in x]
plt.plot(x,y5, label = "O($n^2$)")
plt.plot(x,y6, label = "O($2^n$)")
plt.plot(x,y7, label = "O(n!)")
plt.plot(x,y8, label = "O($n^n$)")
plt.legend()
plt.title('Big O Notation Last four equation')
plt.xlabel('Elements')
plt.ylabel('Operations')


```

```python
x = list(range(1,5))
x2 = list(range(0,4))

y1 = [((0*n) + 1) for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n ** 2 for n in x2]
y6 = [2 ** n for n in x2]
y7 = [math.factorial(n) for n in x2]
y8 = [n ** n for n in x2]

plt.plot(x,y1, label = "O(1)")
plt.plot(x,y2, label = "O(log n)")
plt.plot(x,y3, label = "O(n")
plt.plot(x,y4, label = "O(n log n)")
plt.plot(x,y5, label = "O($n^2$)")
plt.plot(x,y6, label = "O($2^n$)")
plt.plot(x,y7, label = "O(n!)")
plt.plot(x,y8, label = "O($n^n$)")
plt.legend()
plt.title('Big O Notations 1st four equations')
plt.xlabel('Elements')
plt.ylabel('Operations')
```
