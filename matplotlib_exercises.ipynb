```python
import matplotlib.pyplot as plt
```

```python
# Use matplotlib to plot the following equation:
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

x = [0,1,2,3,4,5]
y = [n**2 - n + 2 for n in x]

plt.plot(x,y)
plt.annotate('Orgin', xy=(0, 0), xytext=(1, 6), arrowprops={'facecolor': 'green'})
plt.ylim(0, 15)
plt.xlim(0, 6)
plt.show()

```

```python
#2 Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y = x ** .5   y = x ** 3 y = 2 ** x
# You can use functions from the math module to help implement some of the equations above.

import math

x1 = list(range(0,10))
y1 = [n ** .5 for n in x1]
plt.plot(x1,y1)
plt.title('x^.5)')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x2 = list(range(0,4))
y2 = [n ** 3 for n in x2]
plt.plot(x2,y2)
plt.title('$x^3$')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x3 = list(range(0,3))
y3 = [2 ** n for n in x3]
plt.plot(x3,y3)
plt.title('$2^x$')
plt.xlabel("x")
plt.ylabel("y")
```

```python
#3 Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
plt.plot(x1,y1)
plt.title('$x^{.5}$')
plt.xlabel("x")
plt.ylabel("y")
plt.subplot(1,3,2)
plt.plot(x2,y2)
plt.title('$x^3$')
plt.xlabel("x")
plt.ylabel("y")
plt.subplot(1,3,3)
plt.plot(x3,y3)
plt.title('$2^x$')
plt.xlabel("x")
plt.ylabel("y")
```

```python
#4 Comine the figures you created in the last step into one figure where each of the 4 equations has a 
# different color for the points. Be sure to include a legend.
plt.figure(figsize=(15, 5))
plt.plot(x1, y1, label = 'sqroot x')
plt.plot(x2, y2, label = 'x^3')
plt.plot(x3, y3, label = '2^x')
plt.title('Comparing Three equations')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```
